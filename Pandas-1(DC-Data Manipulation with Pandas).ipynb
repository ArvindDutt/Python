# Defining an IQR function and using it in .agg function's call

```python
def iqr(col):
    return col.quantile(.75)-col.quantile(.25)
```

```python
print((sales_df['temperature_c']).agg(iqr))
```

```python
print((sales_df[['temperature_c','fuel_price_usd_per_l','unemployment']]).agg(iqr))
```

```python
print((sales_df[['temperature_c','fuel_price_usd_per_l','unemployment']]).agg([iqr,np.median]))
```

```python
def percent_change(col):
    return (col-max_fuel_price/max_fuel_price)*100
```

```python
print((sales_df['fuel_price_usd_per_l']).agg(percent_change))
```

# Pivot Tables


In pandas, pivot tables are essentially just another way of performing grouped calculations. That is, the .pivot_table() method is just an alternative to .groupby()

```python

mean_sales_bytype = sales_df.pivot_table(values='weekly_sales',index='type')
print(mean_sales_bytype)
```

```python
mean_med_sales_bytype = sales_df.pivot_table(values='weekly_sales',index='type').agg([np.mean,np.median])

print(mean_med_sales_bytype)
```

```python
mean_sales_bytype_holiday = sales_df.pivot_table(values='weekly_sales',index='type',columns='is_holiday').agg(np.mean)
print(mean_sales_bytype_holiday)
```

### Filling NA values or missing values with some dummy values:

```python
print(sales_df.pivot_table(values='weekly_sales',index='department',columns='type',fill_value=0))
```

#### margins is a shortcut for when you pivoted by two variables, but also wanted to pivot by each of those variables separately: it gives the row and column totals of the pivot table contents.

```python
print(sales_df.pivot_table(values='weekly_sales',index='department',columns = 'type',fill_value=0,margins=True))
```

### Slicing: DataFrames can be sliced by index values or by row/column number

You can only slice an index if the index is sorted (using .sort_index()).
To slice at the outer level, first and last can be strings.
To slice at inner levels, first and last should be tuples.
If you pass a single slice to .loc[], it will slice the rows.

```python
temp_idx.sort_index()
```

```python
#sorting on the rows/ loc of country
temp_idx.sort_index().loc['India':'Russia']
```

```python
temp_idx.sort_index().loc['New Delhi':'Moscow']

# This doesn't work
```

```python
temp_idx.sort_index().loc[("India","New Delhi"):('Russia','Moscow')]

# This is works when we pass in a Tuple of County,City <Start:End>
```

```python
temp_srt_idx = temp_idx.sort_index().loc[("India","Hydrabad"):("Iraq","Baghdad")]
# if I don't use new df over here then the below operations will fail. Try it after deleting the df 'temp_srt_idx'
```

```python
temp_idx.loc[:,'date':'avg_temp_c']
```

```python
temp_srt_idx.loc[("India","Hydrabad"):("Iraq","Baghdad"),'date':'avg_temp_c']

# this is slicing in both directions
```
